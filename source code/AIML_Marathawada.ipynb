## Combine and Harmonize Data

### Subtask:
Concatenate the two datasets into a single DataFrame. Ensure consistent data types and handle initial column alignments, especially for NPK, pH, and yield variables. This step also involves renaming columns for clarity and consistency across the combined dataset.


**Reasoning**:
The subtask requires renaming columns in `df_enhanced` for consistency with `df_marathwada`, calculating a standardized yield column in `df_marathwada` while handling potential division by zero, dropping redundant columns from `df_marathwada`, and then concatenating both dataframes into a single `df_combined` for further processing. This step directly addresses the column alignment and data harmonization requirement.



```python
import numpy as np

# 1. Rename columns in df_enhanced
df_enhanced_renamed = df_enhanced.rename(columns={
    'N': 'Soil_N',
    'P': 'Soil_P',
    'K': 'Soil_K',
    'pH': 'Soil_pH',
    'Crop Yield (kg per hectare)': 'Crop_Yield'
})

# Make a copy to avoid SettingWithCopyWarning later
df_marathwada_processed = df_marathwada.copy()

# 2. For df_marathwada, create a new column named `Crop_Yield_kg_per_hectare`.
# Handle division by zero for 'Area' by replacing 0 with 1
df_marathwada_processed['Area_safe'] = df_marathwada_processed['Area'].replace(0, 1)
df_marathwada_processed['Crop_Yield_kg_per_hectare'] = (df_marathwada_processed['Production'] / df_marathwada_processed['Area_safe']) * 1000

# Replace any resulting inf values with NaN
df_marathwada_processed['Crop_Yield_kg_per_hectare'] = df_marathwada_processed['Crop_Yield_kg_per_hectare'].replace([np.inf, -np.inf], np.nan)

# Drop the temporary 'Area_safe' column
df_marathwada_processed.drop(columns=['Area_safe'], inplace=True)

# 3. Drop original yield-related columns from df_marathwada and rename the new one
df_marathwada_processed.drop(columns=[
    'Crop_Yield', 'Yield_calc', 'StateCrop_Yield', 'State_Yield',
    'Yield_filled', 'StateCrop_ProdMed', 'Crop_ProdMed'
], inplace=True)
df_marathwada_processed.rename(columns={'Crop_Yield_kg_per_hectare': 'Crop_Yield'}, inplace=True)

# 4. Concatenate the processed df_enhanced and df_marathwada into a single DataFrame named df_combined.
df_combined = pd.concat([df_enhanced_renamed, df_marathwada_processed], ignore_index=True, sort=False)

# 5. Display the first 5 rows and the information (.info()) of the df_combined DataFrame
print("First 5 rows of df_combined:")
print(df_combined.head())
print("\nInfo for df_combined:")
df_combined.info()
```

## Prepare Data for Meta-Modeling

### Subtask:
Prepare the combined dataset for meta-modeling by separating the target variable (`Crop_Yield`) from features (`X`), dropping irrelevant columns, handling missing values, identifying categorical and numerical features, applying one-hot encoding, and splitting the data into training, validation, and testing sets.


**Reasoning**:
I need to prepare the combined dataset for meta-modeling. This involves separating the target variable, dropping an irrelevant column, handling missing values in numerical and categorical columns, performing one-hot encoding on categorical features, and finally splitting the data into training, validation, and testing sets. This comprehensive step addresses all the specified data preparation requirements.



```python
from sklearn.model_selection import train_test_split

# 1. Separate the target variable 'Crop_Yield' from df_combined into y, and the remaining columns into X.
y = df_combined['Crop_Yield']
X = df_combined.drop(columns=['Crop_Yield'])

# 2. Drop the 'State_Name' column from X as it contains only one unique value ('Maharashtra').
X = X.drop(columns=['State_Name'])

# Identify numerical columns with missing values
numerical_cols_with_nan = ['Crop_Year', 'Area', 'Production', 'Fertilizer_avg'] + \
                          [col for col in X.columns if col.startswith('Rain_') or col.startswith('Temp_')]

# Filter to include only columns that actually exist in X
numerical_cols_with_nan = [col for col in numerical_cols_with_nan if col in X.columns]

# 3. Fill NaN values in identified numerical columns with their respective column means.
for col in numerical_cols_with_nan:
    if X[col].isnull().any():
        X[col] = X[col].fillna(X[col].mean())

# 4. Identify the categorical column with missing values ('District_Name') and fill NaN with its mode.
if 'District_Name' in X.columns and X['District_Name'].isnull().any():
    mode_val = X['District_Name'].mode()[0]
    X['District_Name'] = X['District_Name'].fillna(mode_val)

# 5. Create a list of categorical features for one-hot encoding.
categorical_features = ['Season', 'Crop', 'District_Name']

# 6. Apply one-hot encoding to the identified categorical features in X using pd.get_dummies().
#    Change drop_first=True to drop_first=False to retain all OHE columns for crop.
X_encoded = pd.get_dummies(X, columns=categorical_features, drop_first=False, dtype=int)

# 7. Split the X_encoded features and the y target variable into training (60%), validation (20%), and testing (20%).
# First split: 80% for training+validation, 20% for testing
X_train_temp, X_test, y_train_temp, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

# Second split: From the training+validation set, take 25% for validation (which is 20% of the original dataset)
X_train, X_val, y_train, y_val = train_test_split(X_train_temp, y_train_temp, test_size=0.25, random_state=42)

# 8. Print the shapes of the split datasets.
print("Shape of X_train:", X_train.shape)
print("Shape of X_val:", X_val.shape)
print("Shape of X_test:", X_test.shape)
print("Shape of y_train:", y_train.shape)
print("Shape of y_val:", y_val.shape)
print("Shape of y_test:", y_test.shape)
```

## Comprehensive Meta-Model Summary

### 1. Handling Variable Inconsistencies Between Datasets

The initial inspection of `Short_enhanced_crop_yield_dataset_maharashtra.csv` (`df_enhanced`) and `Marathwada_Districtwise_Final_Combined_data1.csv` (`df_marathwada`) revealed several inconsistencies, primarily in column naming and the scale of the target variable (`Crop Yield`). These were addressed as follows:

-   **Column Renaming:** Columns in `df_enhanced` related to soil nutrients (`N`, `P`, `K`, `pH`) and yield (`Crop Yield (kg per hectare)`) were renamed to match the conventions of `df_marathwada` (`Soil_N`, `Soil_P`, `Soil_K`, `Soil_pH`, `Crop_Yield`).
-   **Yield Standardization:** The `Crop_Yield` column in `df_marathwada` appeared to be a normalized or relative measure, while `df_enhanced` contained actual yield in kg/hectare. To harmonize this, a new `Crop_Yield` column (in kg/hectare) was calculated for `df_marathwada` by dividing `Production` by `Area` (after handling potential division by zero by replacing 0 area with 1) and multiplying by 1000. Redundant original yield-related columns from `df_marathwada` were then dropped.
-   **Feature Discrepancies:** `df_marathwada` contained additional granular features like `District_Name`, `Crop_Year`, rainfall, and temperature data, which were absent in `df_enhanced`. When combining, these columns naturally resulted in NaN values for `df_enhanced` rows. These missing values were handled during data preparation.

### 2. Meta-Model Architecture

The meta-model employed a stacking ensemble approach. It consists of two main layers:

-   **Base Models:** Three diverse base regression models were chosen to capture different patterns in the data:
    1.  **Random Forest Regressor:** An ensemble tree-based model known for its robustness and ability to handle non-linear relationships.
    2.  **Gradient Boosting Regressor:** Another powerful ensemble tree-based model that builds trees sequentially, correcting errors of previous trees.
    3.  **Simple Neural Network:** A multi-layer perceptron (MLP) with three dense hidden layers (128, 64, 32 neurons, all with ReLU activation) and a single output neuron for regression. This model captures non-linearities through its activation functions.

-   **Meta-Model (Learner):** A simple **Linear Regression** model was used as the meta-learner. Its input features are the predictions generated by the three base models, and its target is the actual `Crop_Yield`.

### 3. Training Process

The training process followed these steps:

1.  **Data Preparation:**

## Interactive Crop Distribution Pie Chart (Attempt 3)

```python
import altair as alt
import ipywidgets as widgets
from IPython.display import display, HTML

# 1. Define a function that generates an Altair pie chart for a given year
def plot_pie_chart_interactive(year):
    # Filter the df_agg DataFrame for the given year
    df_year = df_agg[df_agg['Crop_Year'] == year].copy()

    # Calculate the percentage of 'Area' for each crop
    total_area = df_year['Area'].sum()
    if total_area == 0:
        # Return an empty chart or message if no data for the year
        return alt.Chart({'values': []}).mark_text(
            text=f'No data for Year: {year}'
        ).properties(title=f'Crop Distribution for Year: {year}')

    df_year['Percentage'] = (df_year['Area'] / total_area) * 100

    # Create a base Altair pie chart
    base = alt.Chart(df_year).encode(
        theta=alt.Theta("Area", stack=True)
    ).properties(
        title=f'Crop Distribution for Year: {year}'
    )

    # Draw the arcs (pie slices)
    pie = base.mark_arc(outerRadius=120).encode(
        color=alt.Color("Crop"),
        order=alt.Order("Area", sort="descending"),
        tooltip=[
            alt.Tooltip("Crop"),
            alt.Tooltip("Area"),
            alt.Tooltip("Percentage", format=".1f")
        ]
    )

    # Create text labels for the pie slices
    text = base.mark_text(radius=140).encode(
        text=alt.Text("Percentage", format=".1f"),
        order=alt.Order("Area", sort="descending"),
        color=alt.value("black")  # Set text color to black
    )

    # Combine the arc chart and the text labels
    chart = pie + text

    # Save the chart as an HTML file as a fallback for display issues (optional, but good practice)
    file_name = f'crop_distribution_{year}.html'
    chart.save(file_name)
    print(f"Saved chart for year {year} to {file_name}")

    return chart

# 2. Determine the minimum and maximum crop years from df_agg
min_year = int(df_agg['Crop_Year'].min())
max_year = int(df_agg['Crop_Year'].max())

# 3. Create an ipywidgets.IntSlider for year selection
year_slider_interact = widgets.IntSlider(
    value=min_year,
    min=min_year,
    max=max_year,
    step=1,
    description='Crop Year:',
    continuous_update=False  # Only update on release
)

# 4. Use ipywidgets.interact to link the slider to the function directly
# This method typically displays the widget directly.
widgets.interact(plot_pie_chart_interactive, year=year_slider_interact);
```

### How Climate Change Coefficients Can Be Used to Adjust Future Climate Variable Predictions

The climate change coefficients calculated through linear regression represent the average annual change (increase or decrease) for each climate variable over the observed period (`Crop_Year`). These coefficients can be directly applied to forecast future climate conditions.

**Methodology for Future Predictions:**

1.  **Baseline Value:** Start with a recent average value for each climate variable (e.g., the mean value from the last few years available in `df_climate_yearly_avg`). This serves as the `current_year_baseline`.
2.  **Projection Year:** Define the future year for which a prediction is desired (e.g., `future_year = 2030`).
3.  **Years to Project:** Calculate the number of years from the baseline year to the future year (`years_to_project = future_year - current_year`).
4.  **Adjusting Predictions:** For each climate variable, the future predicted value can be estimated using the following formula:

    `Future_Value = Current_Year_Baseline + (Change_Coefficient * Years_to_Project)`

**Example:**

If the `Rain_ANNUAL` change coefficient is -9.5333 and the current annual rainfall baseline is 700 mm (for a baseline year of 2014), to predict rainfall for 2030 (16 years into the future):

`Future_Rain_ANNUAL = 700 + (-9.5333 * 16) = 700 - 152.5328 = 547.4672 mm`

This indicates a projected decrease in annual rainfall by 2030.

**Implications and Considerations:**

*   **Rate of Change:** The coefficient directly quantifies the average rate of change per year. A positive coefficient indicates an increasing trend, while a negative coefficient indicates a decreasing trend.
*   **Linear Assumption:** This method assumes a linear trend in climate variables, which may not always hold true over long periods. Climate change can be non-linear, and more sophisticated models might be needed for highly accurate long-term forecasts.
*   **Uncertainty:** These are point estimates and do not account for the inherent uncertainty in climate projections. Incorporating confidence intervals from the linear regression could provide a range of possible future values.
*   **Decision Making:** Even with these assumptions, such coefficients provide valuable insights for agricultural planning, allowing farmers and policymakers to anticipate changes in conditions like water availability (rainfall), growing season length (temperature), and soil fertility (nutrients/pH).
*   **Integration with Predictive Models:** These adjusted future climate variable predictions can then be fed into crop yield prediction models (like the meta-model developed earlier) to forecast future crop yields under changing climate conditions.

```python
import numpy as np

# 1. Create X_prediction_template from X_base_template and adjusted_farmer_inputs
X_prediction_template = X_base_template.copy()

# Update climate variables in X_prediction_template using values from adjusted_farmer_inputs
X_prediction_template['Rain_ANNUAL'] = adjusted_farmer_inputs['Annual Rainfall']
X_prediction_template['Temp_Annual_Min'] = adjusted_farmer_inputs['Annual Minimum Temperature']
X_prediction_template['Temp_Annual_Max'] = adjusted_farmer_inputs['Annual Maximum Temperature']
X_prediction_template['Soil_N'] = adjusted_farmer_inputs['Soil_N']
X_prediction_template['Soil_P'] = adjusted_farmer_inputs['Soil_P']
X_prediction_template['Soil_K'] = adjusted_farmer_inputs['Soil_K']
X_prediction_template['Soil_pH'] = adjusted_farmer_inputs['Soil pH']
X_prediction_template['Fertilizer_avg'] = adjusted_farmer_inputs['Fertilizer_avg']

# Update Crop_Year
X_prediction_template['Crop_Year'] = adjusted_farmer_inputs['Present Year']

# Handle District_Name one-hot encoding
# First, set all District_Name_ columns to 0
all_ohe_district_cols = [col for col in X_prediction_template.index if col.startswith('District_Name_')]
X_prediction_template[all_ohe_district_cols] = 0.0

# Then set the specific District_Name_ for the farmer's input to 1
farmer_district_ohe_col = f"District_Name_{adjusted_farmer_inputs['District_Name']}"
if farmer_district_ohe_col in X_prediction_template.index:
    X_prediction_template[farmer_district_ohe_col] = 1.0
else:
    print(f"Warning: Farmer's district '{adjusted_farmer_inputs['District_Name']}' not found in OHE columns.")

# Ensure all Crop_ OHE columns are 0 (they should be from X_base_template already, but for safety)
all_ohe_crop_cols = [col for col in X_prediction_template.index if col.startswith('Crop_')]
X_prediction_template[all_ohe_crop_cols] = 0.0


# 2. Initialize dictionaries to store crop-specific yield sensitivities and CYCs
crop_yield_sensitivities = {}
crop_yield_coefficients_farmer = {}

# Get CCC values for the key climate variables
ccc_values = {
    var: climate_coefficients.get(var, 0.0)
    for var in key_climate_vars_for_cyc
}

# 3. Iterate through each of the dominant_crops_top15
for crop in dominant_crops_top15:
    # 4a. Create a copy of the X_prediction_template and activate the one-hot encoded column for the current crop
    X_current_crop_base = X_prediction_template.copy()
    crop_ohe_col_name = f'Crop_{crop}'

    if crop_ohe_col_name in X_current_crop_base.index:
        X_current_crop_base[crop_ohe_col_name] = 1.0
    else:
        # Skip this crop if its OHE column is not found (e.g., if it was not in X_encoded)
        # This should no longer happen for dominant_crops_top15 with drop_first=False.
        print(f"Warning: OHE column for '{crop}' not found. Skipping CYC calculation for this crop.")
        continue

    # 4b. Predict the baseline yield for this crop using the meta-model
    X_base_input_df = pd.DataFrame([X_current_crop_base], columns=X_encoded.columns)

    rf_pred_base = rf_model.predict(X_base_input_df)
    gb_pred_base = gb_model.predict(X_base_input_df)
    nn_pred_base = nn_model.predict(X_base_input_df).flatten()

    X_meta_input_base = np.column_stack((rf_pred_base, gb_pred_base, nn_pred_base))
    baseline_yield_crop = meta_model.predict(X_meta_input_base)[0]

    current_crop_sensitivities = {}
    cyc_crop_val = 0.0

    # 4c. For each of the key_climate_vars_for_cyc
    for var in key_climate_vars_for_cyc:
        ccc_val = ccc_values[var]

        # If CCC is 0, sensitivity is effectively 0, avoid division by zero
        if ccc_val == 0.0:
            current_crop_sensitivities[var] = 0.0
            continue

        # 4c.i. Create a copy of the current crop's X_prediction_template
        X_perturbed_input_crop = X_current_crop_base.copy()

        # 4c.ii. Perturb the specific climate variable by adding its corresponding CCC
        X_perturbed_input_crop[var] += ccc_val

        # Convert to DataFrame for model prediction
        X_perturbed_input_df = pd.DataFrame([X_perturbed_input_crop], columns=X_encoded.columns)

        # 4c.iii. Predict the perturbed yield using the meta-model
        rf_pred_perturbed = rf_model.predict(X_perturbed_input_df)
        gb_pred_perturbed = gb_model.predict(X_perturbed_input_df)
        nn_pred_perturbed = nn_model.predict(X_perturbed_input_df).flatten()

        X_meta_input_perturbed = np.column_stack((rf_pred_perturbed, gb_pred_perturbed, nn_pred_perturbed))
        perturbed_yield_crop = meta_model.predict(X_meta_input_perturbed)[0]

        # 4c.iv. Calculate the yield sensitivity (Yield/X)
        d_yield_d_X = (perturbed_yield_crop - baseline_yield_crop) / ccc_val
        current_crop_sensitivities[var] = d_yield_d_X

        # 4c.v. Accumulate the product of this sensitivity and the CCC into the crop's total CYC
        cyc_crop_val += d_yield_d_X * ccc_val

    # 4d. Store the calculated CYC for the current crop
    crop_yield_sensitivities[crop] = current_crop_sensitivities
    crop_yield_coefficients_farmer[crop] = cyc_crop_val

# # 5. Print the calculated Crop Yield Coefficients for all top 15 crops
print("\nCrop Yield Coefficients (CYC) for farmer's inputs:")
for crop, cyc in crop_yield_coefficients_farmer.items():
    print(f"  {crop}: {cyc:.4f}")
```

## Calculate Crop Sustainability Coefficient (CSC)

### Subtask:
Calculate the Crop Sustainability Coefficient (CSC) for each crop using the formula: CSC_crop =  * Normalized(CYC_crop) + (1 - ) * SS_crop, with =0.6 for a balanced recommendation.


**Reasoning**:
The previous code failed because 'Crop' was the index of `df_cyc_normalized`, not a column. To resolve this, I need to reset the index of `df_cyc_normalized` so that 'Crop' becomes a regular column before attempting to merge or select it.



```python
from sklearn.preprocessing import MinMaxScaler

# 1. Define alpha () to control the trade-off between yield and sustainability.
#    As per instructions:  = 0.6 for a balanced recommendation.
alpha = 0.6

# 2. Normalize the CYC_Value column in df_cyc_farmer_sorted
#    Create a copy to avoid modifying the original DataFrame until ready for merge
df_cyc_normalized = df_cyc_farmer_sorted.copy()

# Initialize the MinMaxScaler
scaler = MinMaxScaler(feature_range=(0, 1))

# Reshape the data for scaling (MinMaxScaler expects 2D array)
cyc_values_reshaped_adj = df_cyc_normalized['CYC_Value'].values.reshape(-1, 1)

# Fit and transform the CYC values
df_cyc_normalized['Normalized_CYC'] = scaler.fit_transform(cyc_values_reshaped_adj)

# Reset index to make 'Crop' a column for merging
df_cyc_normalized = df_cyc_normalized.reset_index()

# 3. Merge df_cyc_normalized (which now contains Normalized_CYC) with df_csi (which contains Sustainability_Score)
#    Ensure df_csi has 'Crop' as a column, or reset its index.
df_csi_reset = df_csi.reset_index()
df_csi_reset.rename(columns={'index': 'Crop'}, inplace=True)

merged_final_df = pd.merge(
    df_cyc_normalized[['Crop', 'CYC_Value', 'Normalized_CYC']],
    df_csi_reset[['Crop', 'Sustainability_Score']],
    on='Crop',
    how='left'
)

# 4. Calculate the Final_Score (which represents the CSC) for each crop.
#    CSC_crop =  * Normalized(CYC_crop) + (1 - ) * SS_crop
merged_final_df['CSC'] = alpha * merged_final_df['Normalized_CYC'] + \
                           (1 - alpha) * merged_final_df['Sustainability_Score']

# 5. Print the resulting DataFrame, sorted by Final_Score (CSC) in descending order.
print("\nDataFrame with calculated CSC (Final_Score):")
print(merged_final_df.sort_values(by='CSC', ascending=False))
```
